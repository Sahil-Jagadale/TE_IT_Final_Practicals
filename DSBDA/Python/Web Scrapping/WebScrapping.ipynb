```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
URL = "https://www.amazon.in/OnePlus-Mirror-Black-128GB-Storage/product-reviews/B07DJHV6VZ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
page = requests.get(URL)
```

```python
page
```

```python
soup = bs(page.content,'html.parser')
print(soup.prettify())
```

```python
#Extracting Customer names
names = soup.find_all('span',class_='a-profile-name')
cust_name = []
for i in range(0,len(names)):
  cust_name.append(names[i].get_text())
cust_name
```

```python
titles = soup.find_all('a',class_='review-title-content')
titles
review_title = []
for i in range(0,len(titles)):
    review_title.append(titles[i].get_text())
review_title
```

```python
ratings = soup.find_all('span',class_='a-icon-alt')
rates = []
for i in range(0,len(ratings)):
  rates.append(ratings[i].get_text())
rates
```

```python
contents = soup.find_all('span',class_='a-size-base')
review_content = []
for i in range(0,len(contents)):
  review_content.append(contents[i].get_text())
review_content
```

```python
import pandas as pd

df = pd.DataFrame()
cust_name = cust_name[:8]
review_title = review_title[:8]
rates = rates[:8]
review_content = review_content[:8]
df['Customer Name']=cust_name
df['Review Title']=review_title
df['Ratings'] = rates
df['Reviews'] = review_content
df["Numerical Ratings"] = df["Ratings"].str.extract(r'(\d+\.\d+|\d+)').astype(float)
df
```

```python
df_sorted = df.sort_values(by="Numerical Ratings", ascending=False)

# Display the sorted DataFrame
print(df_sorted)
```
