```python
def outlier_remover(data,threshold=3):
  num = data.select_dtypes(include=['int64','float64'])
  zscore = np.abs((num-num.mean())/num.std())
  out = zscore > 3
  data = num[~out.any(axis=1)]
  return data
df = outlier_remover(df)
```

```python
#Another method for oultier removal
def rem2(df):
  df = df.select_dtypes(include=['number'])
  Q1 = df.quantile(0.25)
  Q2 = df.quantile(0.75)
  IQR = Q2 - Q1;
  lower_threshold = Q1 - 1.5 * IQR
  upper_threshold = Q2 + 1.5 * IQR

  df = df[~((df < lower_threshold) | (df > upper_threshold)).any(axis=1)]
  return df
ndf = rem2(df)
df =ndf
```

**Melting Data to Long Format**

The pd.melt() function in pandas is used to transform a DataFrame from wide format to long format. In wide format, each row represents a unique observation, and columns represent different variables or attributes. On the other hand, in long format, each row represents a single observation for a particular variable, with additional columns specifying the variable name and its corresponding value.

```python
df_melted = pd.melt(df, id_vars=['X','Y'], var_name="Metrics", value_name='values')
df_melted
```

**Casting data to wide format**

```python
df_casted = df_melted.pivot_table(index=['X','Y'],columns='Metrics',values='values')
df_casted
```
