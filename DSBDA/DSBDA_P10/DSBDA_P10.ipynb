```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Iris.csv",header=None)
```

```python
df.head()
```

```python
df.columns = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']
df.head()
```

```python
df.isnull().sum()
```

```python
df['Species'].unique()
```

**Create data subsets for different species**

```python
df_setosa = df[df['Species']=='Iris-setosa']
df_setosa.head()
```

```python
df_versicolor = df[df['Species']=='Iris-versicolor']
df_versicolor.head()
```

```python
df_virginica = df[df['Species']=='Iris-virginica']
df_virginica.head()
```

**Merge two Subsets**

```python
df_merge = pd.concat([df_setosa,df_versicolor])
df_merge
```

**Sort Data by Petal Length**

```python
df_sort = df.sort_values('PetalLengthCm',ascending=True)
df_sort.head()
```

**Transposing Data**

```python
df_transpose = df.transpose()
df_transpose.head()
```

**Melting Data to Long format**

```python
df_melt = pd.melt(df,id_vars=['SepalLengthCm','PetalLengthCm','Species'])
df_melt.head()
```

**Casting data to wide format**

```python
df_cast = pd.pivot_table(df,index=['Species'])
df_cast.head()
```
