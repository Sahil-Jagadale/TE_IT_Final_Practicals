**Casting**

```python
df_cast_title = pd.pivot_table(df, index=['id', 'original_language'], values='original_title', aggfunc='first')
df_cast_languages = pd.pivot_table(df, index=['id', 'original_language'], values='spoken_languages', aggfunc='first')

# Concatenate the two pivot tables along the columns axis
df_cast = pd.concat([df_cast_title, df_cast_languages], axis=1)
```
